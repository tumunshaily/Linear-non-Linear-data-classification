# Creating random data set , Can interpretate as you like

```python
from sklearn.datasets import make_blobs #basic library for making random dataset ( this data set is an matrix)

x,y = make_blobs(n_samples=50,centers=2, random_state = 0 , cluster_std=0.60)
print('x:',x ) # acutal data points 
print()
print('y:',y) # data point variance ( check the center variable)
```

# scatter plot

```python
plt.scatter(x[:,0],x[:,1],c=y , s=50 , cmap='summer'); 

```

# Hypothesis plot in our scatter plot 


```python
xfit = np.linspace(-1,3.5) 
plt.scatter(x[:,0],x[:,1],c=y , s=50,cmap='summer')


for m,c in [(1,0.65),(0.5,1.6),(-0.2,2.9)]:  #values of m,b are found manually
    plt.plot(xfit,m*xfit+c , '-k')          # Y= M*x + C this is used 
    plt.xlim(-1,3.5)
```

# Putting margins on our Hypothesis 

```python
xfit = np.linspace(-1,3.5)
plt.scatter(x[:,0],x[:,1],c=y , s=50,cmap='summer')

for m,c,t in [(1,0.65,0.33),(0.5,1.6,0.55),(-0.2,2.9,0.2)]: 
    yfit = m*xfit+c
    plt.plot(xfit,yfit,'-k')
    plt.fill_between(xfit,yfit-t,yfit+t,alpha = 0.5   )

plt.xlim(-1,3.5); #for seeing our polt in the specific range of X
   
```
